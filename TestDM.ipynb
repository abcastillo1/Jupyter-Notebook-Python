## Scikit-learn CountVectorizer

Se utiliza para convertir una colección de documentos de texto en un vector de recuentos de términos.

```python
from sklearn.feature_extraction.text import CountVectorizer
```

## Fit

Realiza el cálculo en los valores de características de los datos de entrada y ajustama este cálculo al transformado

```python
vectorizer = CountVectorizer()
vectorizer.fit(dt2)
print(vectorizer.vocabulary_)
```

## Transform

Se utiliza para codificar el documento

```python
vector = vectorizer.transform(dt2)
```

```python
count_array=vector.toarray()
print(count_array)
```

## Get Feature Names

Imprime los nombres de los términos seleccionados de los documentos sin procesar.

```python
figura3 = pd.DataFrame(data=count_array,columns = vectorizer.get_feature_names())
print(figura3)
```
